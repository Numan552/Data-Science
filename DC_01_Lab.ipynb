```python
#Recursion
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print("Factorial:", factorial(5))
```

```python
#Programming Tasks
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True


number = int(input("Enter a number: "))

if is_prime(number):
    print("Prime Number")
else:
    print("Not Prime")
```

Object-Oriented Programming

```python
class Student:
    pass

s1 = Student()
print(type(s1))
```

```python
name = "Numan"
age = 20

print(name, age)
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Numan", 20)
print(s1.name, s1.age)
```

```python
#OOP Properties
class Student:
    def __init__(self, name):
        self.__name = name   # Private variable

    def get_name(self):
        return self.__name

s = Student("Numan")
print(s.get_name())
```

```python
#OOP Methods
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
print(calc.add(5, 3))
```

```python
#OOP Fundamentals: Inheritance
class Person:
    def speak(self):
        print("I am a person")

class Student(Person):
    def study(self):
        print("I am studying")

s = Student()
s.speak()
s.study()
```

```python
#Exceptions Once Again
class NegativeNumberError(Exception):
    pass

def check_number(num):
    if num < 0:
        raise NegativeNumberError("Negative not allowed")

try:
    check_number(-5)
except NegativeNumberError as e:
    print(e)
```

```python
#Generators
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(5):
    print(num)
```

```python
#Iterators
numbers = [1, 2, 3]
iterator = iter(numbers)

print(next(iterator))
print(next(iterator))
```

```python
#Closures
def outer(x):
    def inner(y):
        return x + y
    return inner

add_five = outer(5)
print(add_five(3))
```
